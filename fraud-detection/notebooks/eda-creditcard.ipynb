```python
# Select only V1 to V28
v_features = [f'V{i}' for i in range(1, 29)]

# Plot histograms
plt.figure(figsize=(20, 20))
for i, col in enumerate(v_features, 1):
    plt.subplot(7, 4, i)  # 7 rows x 4 columns of subplots
    sns.histplot(df[col], bins=50, kde=True, color='skyblue')
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['Amount'], bins=50, kde=True, ax=axes[0], color='salmon')
axes[0].set_title('Transaction Amount Distribution')

sns.histplot(df['Time'], bins=50, kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Transaction Time Distribution')

plt.show()
```

```python
# Count of each class
class_counts = df['Class'].value_counts()
print(class_counts)

# Visualize
plt.figure(figsize=(6, 4))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='pastel')
plt.xticks([0, 1], ['Normal (0)', 'Fraud (1)'])
plt.ylabel('Number of Transactions')
plt.title('Class Distribution')
plt.show()

# Optional: percentage
print("Fraud percentage: {:.4f}%".format(100 * class_counts[1] / class_counts.sum()))
```

```python
# Analyzing Class Imbalance (Run this for both datasets)

def check_imbalance(df, target_col):
    counts = df[target_col].value_counts()
    percentage = df[target_col].value_counts(normalize=True) * 100
    
    print(f"Distribution for {target_col}:")
    print(counts)
    print(f"\nPercentage of Fraud: {percentage[1]:.2f}%")
    
    # Visualization
    sns.countplot(x=target_col, data=df)
    plt.title(f'Class Distribution: {target_col}')
    plt.show()

# Example Usage:
# check_imbalance(fraud_df, 'class')
# check_imbalance(credit, 'Class')
```
